# Data Preprocessing

```python
import pandas as pd
```

```python
objects = pd.read_csv('../data/objects.csv')

objects_final = objects[['id', 'status', 'founded_at', 'closed_at', 'country_code', 'investment_rounds', 'invested_companies', 'milestones', 'relationships']]

objects_final.head(5)
```

```python
funding_rounds = pd.read_csv('../data/funding_rounds.csv')

funding_final = funding_rounds[['object_id', 'funded_at', 'funding_round_type', 'raised_amount_usd', 'participants', 'is_first_round', 'is_last_round']]

funding_final.head(5)
```

```python
funding_rounds.shape
```

```python
final = pd.merge(objects_final, funding_final, how='right', right_on='object_id', left_on='id')
```
